```python
import pandas as pd
star_wars=pd.read_csv('StarWars.csv', encoding="ISO-8859-1")
star_wars.drop(index=0, inplace=True)
star_wars.head(10)
```

```python
star_wars.columns     #此語法不用加括號
```

# Data Cleaning

```python
yes_no = {"Yes": True, "No": False}
star_wars['Have you seen any of the 6 films in the Star Wars franchise?']=star_wars\
['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)

star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?']=star_wars\
['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)

print(star_wars.head(10))
```

```python
import numpy as np

#__map__
new_content={'Star Wars: Episode I  The Phantom Menace':True, \
          'Star Wars: Episode II  Attack of the Clones':True, 
          'Star Wars: Episode III  Revenge of the Sith':True, 
          'Star Wars: Episode IV  A New Hope':True, 
          'Star Wars: Episode V The Empire Strikes Back':True, 
          'Star Wars: Episode VI Return of the Jedi':True, 
          np.nan: False}

for c in star_wars.columns[3:9]:
    star_wars[c]=star_wars[c].map(new_content)

    
#__rename__
new_name={'Which of the following Star Wars films have you seen? Please select all that apply.':'seen_1', \
        'Unnamed: 4':'seen_2', \
        'Unnamed: 5':'seen_3', \
        'Unnamed: 6':'seen_4', \
        'Unnamed: 7':'seen_5', \
        'Unnamed: 8':'seen_6'}

star_wars.rename(new_name, axis=1, inplace=True)
    

#__print__
star_wars.head()
```

```python
#__float__
star_wars.iloc[:, 9:15]=star_wars.iloc[:, 9:15].astype(float)



#__rename__
to_rename={'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.':\
           'ranking_1', 'Unnamed: 10':'ranking_2',\
           'Unnamed: 11':'ranking_3', 'Unnamed: 12':'ranking_4', 'Unnamed: 13':'ranking_5', 'Unnamed: 14':'ranking_6'}

star_wars.rename(to_rename, axis=1, inplace=True)


#__print__
star_wars.head(10)
```

# Ranking of Audience's Preference

```python
star_wars.iloc[:, 9:15].mean()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

star_wars.iloc[:, 9:15].mean().plot.bar()
plt.xticks(rotation=45, labels=['epi_1', 'epi_2', 'epi_3', 'epi_4', 'epi_5', 'epi_6'], ticks=[0, 1, 2, 3, 4, 5])
plt.title('Average Ranking of Six Episodes')
```

It looks like the "original" movies are rated much more highly than the newer ones.

# Ticketbox

```python
star_wars.iloc[:, 3:9].sum()
```

```python
star_wars.iloc[:, 3:9].sum().plot.bar()
plt.xticks(rotation=45, labels=['epi_1', 'epi_2', 'epi_3', 'epi_4', 'epi_5', 'epi_6'], ticks=[0, 1, 2, 3, 4, 5])
plt.title('Average Ticketbox of Six Episodes')
```

It appears that the original movies were seen by more respondents than the newer movies. This reinforces what we saw in the rankings, where the earlier movies seem to be more popular.

# The Differences between Gender on Data

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

```python
plt.figure(figsize=(14, 3.5))

#__male__
plt.subplot(1, 2, 1)
males.iloc[:, 9:15].mean().plot.bar(color='#088C8C')
plt.xticks(rotation=45, labels=['epi_1', 'epi_2', 'epi_3', 'epi_4', 'epi_5', 'epi_6'], ticks=[0, 1, 2, 3, 4, 5])
plt.title('Average Ranking of Six Episodes for Male')


#__female__
plt.subplot(1, 2, 2)
females.iloc[:, 9:15].mean().plot.bar(color='#F28F16')
plt.xticks(rotation=45, labels=['epi_1', 'epi_2', 'epi_3', 'epi_4', 'epi_5', 'epi_6'], ticks=[0, 1, 2, 3, 4, 5])
plt.title('Average Ranking of Six Episodes for Female')
```

```python
plt.figure(figsize=(14, 3.5))

#__male__
plt.subplot(1, 2, 1)
males.iloc[:, 3:9].sum().plot.bar(color='#088C8C')
plt.xticks(rotation=45, labels=['epi_1', 'epi_2', 'epi_3', 'epi_4', 'epi_5', 'epi_6'], ticks=[0, 1, 2, 3, 4, 5])
plt.title('Average Ticketbox of Six Episodes for Male')


#__female__
plt.subplot(1, 2, 2)
females.iloc[:, 3:9].sum().plot.bar(color='#F28F16')
plt.xticks(rotation=45, labels=['epi_1', 'epi_2', 'epi_3', 'epi_4', 'epi_5', 'epi_6'], ticks=[0, 1, 2, 3, 4, 5])
plt.title('Average Ticketbox of Six Episodes for Female')
```

Interestingly, more males watches episodes 1-3, but males liked them far less than females did.
